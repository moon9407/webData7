```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
from bs4 import BeautifulSoup
import time
import os
import random
```

```python
url = "https://www.melon.com/"
browser = webdriver.Chrome() # 창열기
browser.maximize_window() # 최대창 확대
# 1. naver 페이지 열기
browser.get(url)
```

```python
elem = browser.find_element(By.XPATH,'//*[@id="gnb_menu"]/ul[1]/li[1]/a')
time.sleep(1)
elem.click()
```

```python
### html 소스를 Beautifulsoup로 변경
soup = BeautifulSoup(browser.page_source,"lxml")
forms = soup.find("form",{"id":"frm"})
tables = forms.find("table")
trs = tables.find_all("tr")
tds = trs[1].find_all("td") # 상단 타이틀도 함께 넘어옴

```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"}
##### 100번 반복
for i,tr in enumerate(trs):
    if i==0: continue
    tds = tr.find_all("td")
    # 순위
    rank = tds[1].find("span",{"class":"rank"}).text.strip()
    # 이미지
    img = tds[3].find("img")['src']
    os.makedirs("melon",exist_ok=True) # 폴더가 없을때만 생성됨.
    img_url = requests.get(img,headers=headers)
    with open(f"melon/melon_{i}.jpg","wb") as f:
        f.write(img_url.content)
    # 노래제목,가수
    song_info1 = tds[5].find_all("span")[0].text.strip()
    song_info2 = tds[5].find_all("span")[1].text.strip()
    # 좋아요 개수
    cnt =  tds[7].find("span",{"class":"cnt"}).text.strip().split('\n')[1].replace(",","")
    cnt = int(cnt)
    print(rank,img,song_info1,song_info2,cnt)


```
